```python
from trainer import QLearn
import gymnasium as gym

EPISODES = 10000

env = gym.make("Taxi-v3", render_mode="rgb_array")

q_learn = QLearn(env, alpha=1e-1, gamma=0.95, epsilon=1, epsilon_decay=0.9995, min_epsilon=0.01)
rewards = q_learn.train(episodes=EPISODES, stats_interval=500)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def moving_average(data, window_size):
    weights = np.ones(window_size) / window_size
    return np.convolve(data, weights, mode='valid')

plt.figure()
plt.plot(moving_average(rewards, int(EPISODES/10)))
```
